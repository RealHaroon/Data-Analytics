# Series
    - One dimensional labeled array
    - Has Exactly one dtype
    - Syntanx pd.Series(["data"],index=["Optional"])


```python
import pandas as pd
```

```python
pd.Series([1,2,3,4])
pd.Series([1,2,3,4],index=['a','b','c','d'])
```

```python
series=pd.Series([1,2,3,4],index=['a','b','c','d'],dtype="float",name="PythonSeries")
series
```

```python
type(series)
```

```python
series.dtype
```

```python
series=pd.Series(12,index=[1,2,3,4])
series
```

### Creating A Series Using Dictionary

```python
dic={"Name":["Haroon","Abbas"],"Age":[20,32]}
series=pd.Series(dic)
series
```

## Arithmetic Operaitons on Series

```python
series1=pd.Series([10,20,30])
series2=pd.Series([30,20,10])
```

```python
series1+series2
```

```python
series1-series2
```

```python
series1*series2
```

```python
series1/series2
```
