# Series
    - One dimensional labeled array
    - Has Exactly one dtype
    - Syntanx pd.Series(["data"],index=["Optional"])


```python
import pandas as pd
```

```python
pd.Series([1,2,3,4])
pd.Series([1,2,3,4],index=['a','b','c','d'])
```

# DataFrame
    - Two Dimensional Data structure containing rows and columns, like table
    - A colleciton of Series
    - Each column is a series
    - Syntax: pd.DataFrame({"Name":["Haroon","Ali","John"]})
    - Columns can have different dtypes

```python
pd.DataFrame({ "Name":["Haroon","Ali","John"],"Age":[20,32,12]})
```
