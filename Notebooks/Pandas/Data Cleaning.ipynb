## DropNA
    dropna(): Drops Rows that contains NaN Values
    dropna(axis=1) : Drops Columns that contains NaN Values

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

##### __how= Parameter using to check if row contains NaN__
    "any" : checks if row contains ANY NaN
    "all" : checks if row is full of NaN  

```python
df.dropna(how="any")
```

```python
df.dropna(how="all")
```

##### __subset=["ColName"] : removes NaN from Particular columns__

```python
df.dropna(subset=['Age'])
```

##### __inplace=True : save changes into Original DataFrame__
    

```python
df.dropna(inplace=True)
df
```

```python
df
```

##### __thresh= n : removes rows containing n number or NaN__

```python
df.dropna(thresh=1)
```

## FillNa

```python
df
```

### fillna(data) : to fill NaN values with given data.
    fillna({ 'ColName': Data }) : to fill data in particular column

```python
df.fillna('python')
```

```python
df.fillna({"Age":0,"Experience":0})
```

### Method Parameter: Used to fill data of last row or next row
    method="ffill" : For Forward Filling (fill data same as previous row)
    method="bfill" : For Backward Filling (fill data same as next row)

#### __obj.ffill() & obj.bfill__ can also be used instead.

```python
# df.fillna(method="ffill")
df.ffill()
```

```python
# df.fillna(method="bfill")
df.bfill()
```

```python
df
```

### __axis=1 : for filling data of Columns__

```python
# f-fill
# df.fillna(method='ffill',axis=1)
# df.ffill(axis=1)

#b-fill
# df.fillna(method='bfill',axis=1)
# df.bfill(axis=1)
```

### Limit Parameter
    limit = n : fill data into column upto n limit

```python
df.fillna('python',limit=1)
```

## Replace

```python
df=pd.read_csv('D:\\MyEnvironment\\Jupyter-Notebook\\Notebooks\\Dataset\\Data.csv')
df
```

### Replace function  : to replace any/all value within data set
    df.replace(to_replace='Ali', value='Raza')
    to_replace= : data to be replaced
    value= : replacement value


```python
# df.replace(to_replace='Ali',value='Raza')
df.replace(to_replace=[1,2,3,4,5,6,7],value=20)
```

### Regex Parameter

```python
df.replace('[A-Za-z]','python',regex=True)
```

### Replace Data of Partcular Column Using Dictionary

```python
df.replace({"Name":'[A-Z]'},25,regex=True)
```

### Replace Data using Forward Filling and Backward Filling

### __Forward Filling__

```python
df.replace(1,method='ffill')
```

### __Backward Filling__

```python
df.replace(1,method='bfill')
```

### Limit Parameter
    limit = n : fill data into column upto n limit

```python
df.replace('IT',method="bfill",limit=1)
```

### Inplace Parameter 
     Save changes into original dataset

```python
df.replace('IT',method="bfill",limit=1,inplace=True)
```
