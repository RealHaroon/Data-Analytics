# **`count()` — Counts NON-NULL values**

```python
import pandas as pd
```

### **On a Series**

```python
s=pd.Series([10,20,30,None,40])
s.count()
```

### **On a DataFrame --Column-wise non-null counts**

```python
df=pd.DataFrame({ 'Name':['Ali','Sara','Haroon'],'Age':[20,30,None]})
df.count()
```

```python
df.count(axis=1)
```

# **`value_counts`— Counts FREQUENCY of values**

### **On a Series**

```python
s=pd.Series(['IT','HR','IT','SE','SE',None])
s.value_counts()
```

### **On a DataFrame**

```python
df=pd.DataFrame({ 'Name':['Ali','Ali','Sara','Haroon'],'Age':[20,30,None,21]})
df.value_counts() # Wrong use
df['Name'].value_counts()
```

### **Include NaN**

```python
s.value_counts(dropna=False)
```

### **Normalize (Percentages)**

```python
s.value_counts(normalize=True)
```
