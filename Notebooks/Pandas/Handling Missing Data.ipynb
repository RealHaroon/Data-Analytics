## DropNA
    dropna(): Drops Rows that contains NaN Values
    dropna(axis=1) : Drops Columns that contains NaN Values

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

##### How Parameter using to check if row contains NaN
    how=
    "any" : checks if row contains ANY NaN
    "all" : checks if row is full of NaN  

```python
df.dropna(how="any")
```

```python
df.dropna(how="all")
```

##### Subset Parameter removes NaN from Particular columns
        subset=['ColName']

```python
df.dropna(subset=['Age'])
```

##### Inplace Parameters save changes into Original 
    inplace=True
    

```python
df.dropna(inplace=True)
df
```

#####  Thresh Parameter removes rows containing n number or NaN__
        thresh= n

```python
df.dropna(thresh=1)
```

 # Pandas interpolate() Methods

| Method | Description | When to Use |
|------|------------|------------|
| `linear` | Fills NaN using straight-line interpolation between values (default) | General numeric data |
| `time` | Interpolates based on time index spacing | Time-series data with DatetimeIndex |
| `index` | Uses index values as x-coordinates | When index represents meaningful numeric order |
| `values` | Uses actual data values for interpolation | Rare, numeric-only cases |
| `nearest` | Fills NaN using nearest non-NaN value | When closest value matters |
| `zero` | Step-wise interpolation using zero-order spline | Discrete step data |
| `slinear` | Linear spline interpolation | Smoother than linear |
| `quadratic` | Second-order spline interpolation | Curved numeric trends |
| `cubic` | Third-order spline interpolation | Smooth curves (advanced) |
| `barycentric` | Polynomial interpolation using barycentric method | Mathematical datasets |
| `polynomial` | Polynomial interpolation (requires `order` parameter) | Controlled curve fitting |
| `spline` | Spline interpolation (requires `order`) | Advanced numerical data |
| `pchip` | Shape-preserving cubic interpolation | Avoids overshooting |
| `akima` | Akima spline interpolation | Noisy or uneven data |
| `pad` / `ffill` | Forward fill (previous value) | Sequential data |
| `backfill` / `bfill` | Backward fill (next value) | Reverse sequential data |



### Limit Parameter
    limit = n : fill data into column upto n limit

```python
df.interpolate(limit=1)
```

### Limit Direction Parameter
        Direction in which limit applies.
    

```python
df.interpolate(limit_direction='backward',limit=1)
```

### Limit Area Parameter 
         Area in which limit applies

```python
df.interpolate(limit_area='outside')
```

### Inplace Parameter
