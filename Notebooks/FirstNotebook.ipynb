#### loc & iloc used for selecting rows & columns


**Loc**  :
* Labed-based indexing,
* It'll give you rows and cols by name,
* It includes both start and end in slicing ,

```python
df=pd.DataFrame({"Name":["haroon","talha","raimal"], "age":[23,42,12]},index=["a","b","c"])
```

```python
# loc Demo
df.loc["b"]
```

```python
df.loc["a":"c"]
```

**Iloc**:
* Indexing by number
* It follows 0 indexing
