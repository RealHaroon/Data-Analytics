Q No.1.A Write a Python program to remove all occurrences of the value 10 from the list:
numbers = [10, 5, 10, 20, 30, 10, 40]. Output the updated list.

```python
numbers=[10,5,20,30,10,40]
for i in numbers:
    if i== 10:
        numbers.remove(i)
print(numbers)
```

Write a Python program that takes a list of integers and creates a new list that contains only the unique elements in ascending order.
Example input: [4, 2, 7, 2, 9, 4, 1]. Expected output: [1, 2, 4, 7, 9]

```python
def removeDuplicate(lst):
    unique_vals=set(lst)
    new_list=list(unique_vals)
    return new_list

lst=[5,2,6,2,5,4,1]
removeDuplicate(lst)
```


Q No.2.B
Write a Python program to count the frequency of each word in the list and store the result in a dictionary.
Example list: words = [''apple",
", "banana", "apple", "orange",
", "banana", "apple"]
Expected dictionary: {"apple": 3, "banana": 2, "orange": 1}

```python
words = ["apple","banana", "apple", "orange","banana", "apple"]
freq={}
for i in words:
    if i not in freq:
        freq.update({i:1})
    else:
        freq.update({i:freq[i]+1})
print(freq)
```
